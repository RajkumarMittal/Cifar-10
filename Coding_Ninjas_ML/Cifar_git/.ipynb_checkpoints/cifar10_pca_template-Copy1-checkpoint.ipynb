```python
import cifar10
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import time
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
cifar10.data_path = "data/CIFAR-10/"
```

```python
cifar10.maybe_download_and_extract()
```

```python
class_names = cifar10.load_class_names()
class_names
```

```python
images_train, cls_train, labels_train = cifar10.load_training_data()
images_test, cls_test, labels_test = cifar10.load_test_data()
```

```python
set(labels_train[0])
```

```python
x_train = images_train.reshape(images_train.shape[0],-1)
x_test = images_test.reshape(images_test.shape[0], -1)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
set(cls_train)
```

```python
images_train.shape
```

```python
fig = plt.figure(figsize=(8,8))
for i in range(64):
    ax = fig.add_subplot(8, 8, i+1)
    ax.imshow(images_train[i], cmap= plt.cm.bone)
plt.show()
```

```python
pca = PCA()
pca.fit(x_train)
```

```python
pca.components_.shape
```

```python
pca.explained_variance_.shape
```

```python
k = 0
total = sum(pca.explained_variance_)
currentSum = 0
while(currentSum/total < 0.99):
    currentSum += pca.explained_variance_[k]
    k += 1
k
```

```python
# pca = PCA(n_components=k, whiten=True)
pca = PCA(n_components=k, whiten=False)
x_transform = pca.fit_transform(x_train)
```

```python
x_transform.shape
```

```python
x_approx = pca.inverse_transform(x_transform)
```

```python
x_approx.shape
```

```python
x_approx = x_approx.reshape((50000, 32, 32, 3))
x_approx.shape
```

```python
fig = plt.figure(figsize=(8,8))
for i in range(64):
    ax = fig.add_subplot(8, 8, i+1)
    ax.imshow(x_approx[i], cmap= plt.cm.bone)
plt.show()
```

```python
eigenv = pca.components_
eigenv.shape
```

```python
eigenface = pca.components_.reshape(k, 32, 32, 3)
```

```python
x_train_pca = x_transform
x_test_pca = pca.transform(x_test)
```

```python
x_train_pca.shape
```

```python
x_test_pca.shape
```

```python
from sklearn import svm
```

```python
clf = svm.SVC()
clf.fit(x_train_pca, cls_train)
```

```python
clf.score(x_test_pca, cls_test)
```

```python
y_pred = clf.predict(x_test_pca)
print(classification_report(cls_test,y_pred))
print(confusion_matrix(cls_test,y_pred))
```
